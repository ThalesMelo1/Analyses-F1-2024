```python
import pandas as pd
import numpy as np
import matplotlib as mt
```

```python
DriversDF = pd.read_csv('drivers.csv')
RacesDF = pd.read_csv('races.csv')
ResultsDF = pd.read_csv('results.csv')
```

# Separate only the necessary from the data bases

## Get only the necessary from drivers.csv

```python
DriversDF
```

```python
Drivers = DriversDF[['driverId','forename','driverRef']]
Drivers
```

## Get only the necessary from races.csv

```python
RacesDF
```

```python
Races = RacesDF[['raceId', 'circuitId', 'name', 'date', 'year']].sort_values(by='date', ascending = False)
Races1 = Races[(Races['raceId'] >= 1) & (Races['raceId'] <= 174)]
Races2 = Races[(Races['raceId'] >= 337) & (Races['raceId'] <= 355)]
Races3 = Races[(Races['raceId'] >= 841) & (Races['raceId'] <= 1124)]
RaceOrganizada = pd.concat([Races3, Races2, Races1])
RaceOrganizada
```

```python
RaceOrganizada
RaceOrganizadaCopy = RaceOrganizada
RaceOrganizadaCopy['raceId'] = RaceOrganizadaCopy['date']
RaceOrganizadaCopy
```

## Get only the necessary from results.csv

```python
ResultsDF
```

```python
Results = ResultsDF[['raceId', 'driverId', 'position']]
Results.rename(columns={'raceId':'raceID'}, inplace= True)
Results
```

# Combine all of the information

```python
All = pd.concat([Drivers, RaceOrganizada, Results], axis=1)
All = All.sort_values(by = 'year', ascending=False)
```

```python
Allcopy = All.dropna()
Allcopy
```

```python
Cate = {
    '18': 'Australian Grand Prix',
    '19': 'Malaysian Grand Prix',
    '20': 'Chinese Grand Prix'
}

All['Cate'] = All['raceID'].astype(str).map(Cate)

All[['raceID', 'Cate']]
```
