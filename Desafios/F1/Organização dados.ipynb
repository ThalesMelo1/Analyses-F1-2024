# ORGANIZAÇÃO DOS DADOS

## Importação da biblioteca Pandas e das bases

```python
import pandas as pd
```

```python
DriversDF = pd.read_csv('drivers.csv')
RacesDF = pd.read_csv('races.csv')
ResultsDF = pd.read_csv('results.csv')
```

```python
Drivers = DriversDF[['driverId','forename','driverRef']]
Drivers
```

```python
Races = RacesDF[['raceId', 'circuitId', 'name', 'date', 'year']]
Races
```

```python
Results = ResultsDF[['raceId', 'driverId', 'grid', 'position', 'points', ]]
Results
```

## Criação da Base com elementos selecionados

```python
Resultados = Results.merge(Races, on='raceId', how='left')
```

```python
Resultados = Resultados.merge(Drivers, on='driverId', how='right')
```

```python
Resultados
```

Verificando os tipos de dados

```python
Resultados.dtypes
```

Mudando a coluna position para inteiros 

```python
Resultados['position'] = Resultados['position'].replace({'\\N':999})
```

```python
Resultados['position'] = Resultados['position'].astype(int)
```

Verificando novamente

```python
Resultados.dtypes
```

```python
Resultados = Resultados[['raceId','year','date','name','forename','driverRef','grid','position', 'points']]
Resultados = Resultados.sort_values(by =['date', 'position'], ascending= [False, True])
```

```python
Resultados.to_csv('Base.csv')
```
